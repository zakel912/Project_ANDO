**visualisation data**

```python
import csv
import matplotlib.pyplot as plt


file_names = ["players.csv", "appearances.csv", "transfers.csv", "games.csv", "clubs.csv", "club_games.csv"]


columns_count = []


for file in file_names:
    try:
        with open(file, 'r') as csvfile:
            reader = csv.reader(csvfile)
            headers = next(reader)
            columns_count.append(len(headers))
    except Exception as e:
        print(f"Erreur lors de la lecture de {file}: {e}")
        columns_count.append(0)


plt.figure(figsize=(8, 6))
plt.bar(file_names, columns_count, color='lightblue', edgecolor='black')


plt.title('Number of Columns in Each File', fontsize=14)
plt.xlabel('Files', fontsize=12)
plt.ylabel('Number of Columns', fontsize=12)


for i, count in enumerate(columns_count):
    plt.text(i, count + 0.2, str(count), ha='center', fontsize=10)

plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import prince
import matplotlib.pyplot as plt


appearances = pd.read_csv("appearances.csv")
players = pd.read_csv("players.csv")
transfers = pd.read_csv("transfers.csv")


data = pd.merge(appearances, players, on="player_id")
data = pd.merge(data, transfers[['player_id', 'transfer_fee']], on="player_id", how="left")


data['year'] = pd.to_datetime(data['date']).dt.year


data = data[data['year'] >= 2020]


if 'position' not in data.columns:
    raise ValueError("La colonne 'position' doit être présente dans le fichier players.csv.")


data = data[['position', 'goals', 'assists', 'minutes_played', 'transfer_fee', 'year']]


data = data[data['transfer_fee'].notna() & (data['transfer_fee'] > 0)]


data = data[data['position'] != 'GoalKeeper']

position_weights = {
    'Attack': {'goals': 10, 'assists': 7, 'minutes_played': 3},
    'Midfield': {'goals': 5, 'assists': 8, 'minutes_played': 5},
    'Defender': {'goals': 2, 'assists': 3, 'minutes_played': 8}
}


for position, weights in position_weights.items():
    print(f"\nAnalyse pour le poste : {position}")


    subset = data[data['position'] == position]


    performance_score = (
        subset['goals'] * weights['goals'] +
        subset['assists'] * weights['assists'] +
        (subset['minutes_played'] / 90) * weights['minutes_played']
    )


    performance_score += np.random.uniform(0, 0.01, size=len(subset))


    subset.loc[:, 'performance_score'] = performance_score


    num_categories = min(3, len(subset['performance_score'].unique()))
    subset.loc[:, 'performance_detailed'] = pd.qcut(
        performance_score,
        q=num_categories,
        labels=['Faible', 'Moyen', 'Élevé'][:num_categories],
        duplicates='drop'
    )

    transfer_fee_values = subset['transfer_fee'].dropna()

    num_bins_transfer_fee = min(3, len(transfer_fee_values.unique()))

    if num_bins_transfer_fee > 1:

        try:
            subset.loc[:, 'transfer_fee_detailed'] = pd.qcut(
                subset['transfer_fee'],
                q=num_bins_transfer_fee,
                labels=['Bas', 'Moyen', 'Élevé'][:num_bins_transfer_fee],
                duplicates='drop'
            )
        except ValueError as e:
            print(f"Erreur dans l'utilisation de qcut : {e}")

            subset.loc[:, 'transfer_fee_detailed'] = 'Unique'
    else:

        subset.loc[:, 'transfer_fee_detailed'] = 'Unique'


    print(subset[['performance_score', 'transfer_fee', 'transfer_fee_detailed']].head())


    tables_by_year = {}
    for year in subset['year'].unique():
        year_subset = subset[subset['year'] == year]
        contingency_table = pd.crosstab(
            year_subset['performance_detailed'],
            year_subset['transfer_fee_detailed']
        )


        if contingency_table.shape[0] > 0 and contingency_table.shape[1] > 0:
            tables_by_year[year] = contingency_table


            afc = prince.CA(n_components=2)
            afc = afc.fit(contingency_table)


            row_coords = afc.row_coordinates(contingency_table)
            col_coords = afc.column_coordinates(contingency_table)


            plt.figure(figsize=(10, 7))


            plt.scatter(row_coords[0], row_coords[1], label='Performances', color='blue')
            for i, label in enumerate(contingency_table.index):
                plt.text(row_coords[0][i], row_coords[1][i], label, color='blue')


            plt.scatter(col_coords[0], col_coords[1], label='Transfer Fee', color='red')
            for i, label in enumerate(contingency_table.columns):
                plt.text(col_coords[0][i], col_coords[1][i], label, color='red')


            plt.title(f"AFC - Poste {position} - Année {year}")
            plt.xlabel("Dimension 1")
            plt.ylabel("Dimension 2")
            plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
            plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
            plt.legend()
            plt.grid()
            plt.show()
        else:
            print(f"Table croisée vide pour l'année {year} pour le poste {position}.")
```

```python
import pandas as pd
import numpy as np
import prince
import matplotlib.pyplot as plt

appearances = pd.read_csv("appearances.csv")
players = pd.read_csv("players.csv")
transfers = pd.read_csv("transfers.csv")

data = pd.merge(appearances, players, on="player_id")
data = pd.merge(data, transfers[['player_id', 'transfer_fee']], on="player_id", how="left")

data['year'] = pd.to_datetime(data['date']).dt.year


data = data[data['year'].isin([2021, 2022, 2023])]

if 'position' not in data.columns:
    raise ValueError("La colonne 'position' doit être présente dans le fichier players.csv.")

data = data[['position', 'goals', 'assists', 'minutes_played', 'transfer_fee', 'year']]

data = data[data['transfer_fee'].notna() & (data['transfer_fee'] > 0)]

data = data[data['position'] != 'GoalKeeper']

position_weights = {
    'Attack': {'goals': 10, 'assists': 7, 'minutes_played': 3},
    'Midfield': {'goals': 5, 'assists': 8, 'minutes_played': 5},
    'Defender': {'goals': 2, 'assists': 3, 'minutes_played': 8}
}


fig, axes = plt.subplots(len(position_weights), 3, figsize=(15, 7))

for idx, (position, weights) in enumerate(position_weights.items()):
    print(f"\nAnalyse pour le poste : {position}")

    subset = data[data['position'] == position]

    performance_score = (
        subset['goals'] * weights['goals'] +
        subset['assists'] * weights['assists'] +
        (subset['minutes_played'] / 90) * weights['minutes_played']
    )

    performance_score += np.random.uniform(0, 0.01, size=len(subset))

    subset.loc[:, 'performance_score'] = performance_score

    num_categories = min(3, len(subset['performance_score'].unique()))
    subset.loc[:, 'performance_detailed'] = pd.qcut(
        performance_score,
        q=num_categories,
        labels=['Faible', 'Moyen', 'Élevé'][:num_categories],
        duplicates='drop'
    )

    transfer_fee_values = subset['transfer_fee'].dropna()

    num_bins_transfer_fee = min(3, len(transfer_fee_values.unique()))

    if num_bins_transfer_fee > 1:
        try:
            subset.loc[:, 'transfer_fee_detailed'] = pd.qcut(
                subset['transfer_fee'],
                q=num_bins_transfer_fee,
                labels=['Bas', 'Moyen', 'Élevé'][:num_bins_transfer_fee],
                duplicates='drop'
            )
        except ValueError as e:
            print(f"Erreur dans l'utilisation de qcut : {e}")
            subset.loc[:, 'transfer_fee_detailed'] = 'Unique'
    else:
        subset.loc[:, 'transfer_fee_detailed'] = 'Unique'


    for col_idx, year in enumerate([2021, 2022, 2023]):
        year_subset = subset[subset['year'] == year]
        contingency_table = pd.crosstab(
            year_subset['performance_detailed'],
            year_subset['transfer_fee_detailed']
        )

        if contingency_table.shape[0] > 0 and contingency_table.shape[1] > 0:
            afc = prince.CA(n_components=2)
            afc = afc.fit(contingency_table)

            row_coords = afc.row_coordinates(contingency_table)
            col_coords = afc.column_coordinates(contingency_table)


            ax = axes[idx, col_idx]
            ax.scatter(row_coords[0], row_coords[1], label='Performances', color='blue')
            for i, label in enumerate(contingency_table.index):
                ax.text(row_coords[0][i], row_coords[1][i], label, color='blue')

            ax.scatter(col_coords[0], col_coords[1], label='Transfer Fee', color='red')
            for i, label in enumerate(contingency_table.columns):
                ax.text(col_coords[0][i], col_coords[1][i], label, color='red')

            ax.set_title(f"AFC - Poste {position} - Année {year}")
            ax.set_xlabel("Dimension 1")
            ax.set_ylabel("Dimension 2")
            ax.axhline(0, color='black', linewidth=0.5, linestyle='--')
            ax.axvline(0, color='black', linewidth=0.5, linestyle='--')
            ax.legend()
            ax.grid()
        else:
            print(f"Table croisée vide pour l'année {year} pour le poste {position}.")


plt.tight_layout()
plt.show()
```
